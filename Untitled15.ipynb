```python

#visualize the data
fig,axe = plt.subplots(8,8,figsize = (8,8))
fig.tight_layout(pad = 0.13)
m = X.shape[0]

for i,ax in enumerate(axe.flat):
     random_index = np.random.randint(m)
     ax.imshow(X[random_index].reshape(20,20).T)
     ax.set_title(y[random_index,0])
     ax.set_axis_off()




```

```python
#we got the data set which conatins the number of examples as 1000 and for each image of handwritten
#characters we have the 400 pixels that is 20*20 ( written linearly)
#lets build the model of 3 layers

model = Sequential(
   [
        Input(shape = (400,)),
        Dense(25,activation = "sigmoid",name = "L1"),
        Dense(15,activation = "sigmoid",name = "L2"),
        Dense(1,activation = "sigmoid",name = "L3"),

   ]
)
```

```python
W1,B1 = model.get_layer("L1").get_weights()
```

```python
W1.shape
```

```python
model.compile(optimizer = "adam",loss = "binary_crossentropy")
model.fit(X,y,epochs = 20)
```

```python
X[0].shape
```

```python
#we have built and trained our model successfully on the given data set .
#now we can try to predict with the help of model on any random data point .
random_index = np.random.randint(m)
prediction = model.predict(X[random_index].reshape(1,400))
if prediction > 0.5:
    prediction = 1
else:
    prediction = 0
fig,axe = plt.subplots(1,2,figsize = (8,4))

axe[0].imshow(X[random_index].reshape(20,20).T)
axe[0].set_title(prediction)

axe[1].imshow(X[random_index].reshape(20,20).T)
axe[1].set_title(y[random_index,0])
plt.show()

```
